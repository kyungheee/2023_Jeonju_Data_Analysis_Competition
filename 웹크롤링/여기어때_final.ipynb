```python
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry

import time
import datetime

from bs4 import BeautifulSoup as bs
from datetime import timedelta
from webdriver_manager.chrome import ChromeDriverManager

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException

import pandas as pd
```

```python
# Chrome 옵션 설정
chrome_options = Options()

# 사용자 에이전트 설정
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"
chrome_options.add_argument(f"user-agent={user_agent}")

# Chrome 창 분리 옵션 설정
chrome_options.add_experimental_option('detach', True)

time.sleep(5)

# Chrome 웹 드라이버 시작
driver = webdriver.Chrome(service=Service('./chromedriver'), options=chrome_options)
driver.get('https://www.goodchoice.kr/')

time.sleep(5)
```

```python
# 업소명에 '전주'가 포함된 숙박업소 리스트
search = pd.read_csv('YN_LDGS_PROPERTY_LC_INFO_202304.csv')
search_new = search[search['LDGS_NM'].str.contains('전주')]
search_new
```

```python
# 'LDGS_NM'열데이터를 리스트로 저장
text = search_new['LDGS_NM']
text_list = text.values.tolist()
```

```python
# 호텔이름과 해당 호텔의 href를 저장할 리스트 초기화
hotel_list=[]
href_list=[]
```

```python
# text_list에 있는 호텔들을 여기어때 검색창에 반복적으로 검색
for LDGS in text_list:
    form_element = driver.find_element(By.XPATH,'/html/body/div[1]/header/section/button[2]')
    form_element.click()

    wait = WebDriverWait(driver, 10)
    element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="keyword"]')))

    element.send_keys(LDGS)
    element.send_keys(Keys.ENTER)
        
    wait = WebDriverWait(driver, 5)

    # 검색이 되지 않으면 넘어가고 검색이 된다면 href를 가져옴
    try:
        empty_elements = driver.find_elements(By.CLASS_NAME, 'result_empty')

        if len(empty_elements) > 0:
            keyword_element = empty_elements[0].find_element(By.CLASS_NAME, 'keyword')

        else:
            time.sleep(5)
            url = driver.find_element(By.CSS_SELECTOR, '#poduct_list_area > ul > li > a').get_attribute('href')
            href_list.append(url)

            time.sleep(5)
            hotel = LDGS
            hotel_list.append(hotel)

    except NoSuchElementException:
        pass

    # 쿠키 삭제
    driver.delete_all_cookies()
        
    time.sleep(5)
```

```python
# 호텔이름과 href를 딕셔너리로 href_df에 넣음
data = {'hotel': hotel_list, 'href': href_list}
href_df = pd.DataFrame(data)

# 모든 행과 열을 표시
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# csv파일로 저장
href_df.to_csv('href_data.csv', index=False)
df1= pd.read_csv('href_data.csv')

# 출력 (166개 출력된 게 날아가서 다시 출력했더니 52개밖에 출력이 안됐네용ㅜㅜ 166개 리스트는 미리 저장해뒀습니다.)
display(df1)
```

```python
# 출력된 걸 확인 하고 다시 한번 더 csv파일로 저장
df1.to_csv("여기어때_href.csv",index=False, encoding="utf-8-sig")
```


# review 데이터 프레임 출력하고 저장하기


```python
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry

import time
import datetime

from bs4 import BeautifulSoup as bs
from datetime import timedelta
from webdriver_manager.chrome import ChromeDriverManager

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException

import pandas as pd
```

```python
# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument("--headless")

# 사용자 에이전트 설정
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36"
chrome_options.add_argument(f"user-agent={user_agent}")

# Chrome 창 분리 옵션 설정
chrome_options.add_experimental_option('detach', True)

# Chrome 웹 드라이버 시작
driver = webdriver.Chrome(service=Service('./chromedriver'), options=chrome_options)
driver.get('https://www.goodchoice.kr/')

time.sleep(2)
```

```python
# 데이터를 수집하는 함수 생성
def get_review_data(review_list, hotel):
    
    # 임시로 데이터를 넣을 review_df 초기화
    review_df = []

    # 각 페이지당 리뷰 10개씩 수집 
    for i in range(1, 11):
        try:
            star_element = review_list[0].select_one(f'li:nth-child({i}) > div > div.score_wrap_sm > div.num')
            review_element = review_list[0].select(f'li:nth-child({i}) > div > div.txt')[0]
            date_element = review_list[0].select(f'#review > ul > li:nth-child({i}) > div > span.date')[0]
            
            # 요소가 선택되지 않았을 경우 건너뛰기
            if not star_element or not review_element or not date_element:
                continue
                
            star = star_element.get_text(strip=True)
            review = review_element.get_text(strip=True)
            date = date_element.get_text(strip=True)

            review_df.append([hotel, star, review, date])

        # 에러가 뜨는 경우 건너뛰기    
        except (IndexError, AttributeError, NoSuchElementException):
            continue

    return review_df
```

```python
# 최종적으로 리뷰를 담을 total_review_df 초기화 
total_review_df = pd.DataFrame(columns=['hotel', 'star', 'review', 'date'])
```

```python
# 위에서 생성한 hotel_list, href_list 불러오기
read = pd.read_csv('여기어때_href.csv')
hotel_list = read['hotel'].tolist()
href_list = read['href'].tolist()
```

```python
# href를 통해 각 호텔 사이트 들어가기
for href in href_list:
    
    driver.get(href)
    
    wait = WebDriverWait(driver, 2)
    
    # 팝업창이 뜬다면 닫기 버튼을 누르고 안뜨면 넘어가기
    try:
        popup_element = driver.find_element(By.CLASS_NAME,'msg.popup_cont')
        btn_element = driver.find_element(By.CLASS_NAME,'btn_center')
        btn_element.click()
    
    # 팝업창이 없다면 건너뛰기
    except NoSuchElementException:
        pass

    # 요소를 클릭할 수 있을때까지 기다림
    wait = WebDriverWait(driver, 2)
    element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '#content > div.tab > button.tab_review')))    

    # 리뷰버튼 누르기
    review_button = driver.find_element(By.CSS_SELECTOR, '#content > div.tab > button.tab_review')
    review_button.click()

    time.sleep(2)
    
    

    # 리뷰페이지 1페이지부터 5페이지까지 클릭
    for i in range(1, 6):

        try: 
            page_number_selector = f"#pagination > div > button:nth-child({i})"
            page_number_button = driver.find_element(By.CSS_SELECTOR, page_number_selector)
            page_number_button.click()
        
        # 다음 페이지가 존재하지 않을 때 건너뛰기
        except NoSuchElementException:
            continue

        time.sleep(1)

        # html 파싱
        html = driver.page_source
        soup = bs(html, 'html.parser')
        r = soup.select('#review > ul')

        # 임시로 review_df에 데이터 넣기
        review_df = get_review_data(r, hotel_list[href_list.index(href)])

        # 최종적으로 total_review_df에 review_df을 추가
        total_review_df = pd.concat([total_review_df, pd.DataFrame(review_df, columns=['hotel', 'star', 'review', 'date'])])

        time.sleep(1)

    
    
#     # 다음 페이지 버튼 클릭   
#     next_button = driver.find_element(By.CSS_SELECTOR, "#pagination > div > button.next")
#     next_button.click()

#     time.sleep(2)



#     # 리뷰페이지 6페이지부터 7페이지까지 클릭 (이상하게 8페이지부터 들어가지질 않네요..)
#     for i in range(6, 8):

#         page_number_selector = f"#pagination > div > button:nth-child({i})"
#         page_number_button = driver.find_element(By.CSS_SELECTOR, page_number_selector)
#         page_number_button.click()

#         time.sleep(1)

#         # html 파싱
#         html = driver.page_source
#         soup = bs(html, 'html.parser')
#         r = soup.select('#review > ul')

#         # 임시로 review_df에 데이터 넣기
#         review_df = get_review_data(r, hotel_list[href_list.index(href)])

#         # 최종적으로 total_review_df에 review_df을 추가
#         total_review_df = pd.concat([total_review_df, pd.DataFrame(review_df, columns=['hotel', 'star', 'review', 'date'])])
    
#         time.sleep(1)
```
